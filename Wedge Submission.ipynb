## Reflections
I contracted for a C, and only completed tasks 2 and 3. I did find these tasks easier to tackle once I went over SQL assignments, Dram Shop 1, and the exercise for SQL & Python. I found it easier to build the queries in Google BigQuery first, and then add them to my code to run in Python. I did not have to clean raw data, but I think that if I could do it over again, I would go back and do the project at an A level to gain experience building an idempotent data pipeline. Overall, I think this project is valuable to students because it has real-life application.
